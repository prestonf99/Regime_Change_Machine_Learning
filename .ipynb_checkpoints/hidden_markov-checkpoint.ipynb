## Imports & Data Retrieval

```python
import markov as mk
import pandas as pd
from hmmlearn import hmm
import numpy as np
from pylab import mpl, plt
plt.style.use('seaborn-v0_8')
```

```python
days = 10000 #10,000 Daily Bars
underlying = '^GSPC' #S&P 500
sp = mk.Markov(days, underlying) #Providing Ticker symbol and number of days to background script
```

```python
data = sp.get_data() #Retrieving price data from script
rvol = sp.cc() #Retrieving realized volatility
data = data[29:] #Manually removing NaN's
data['rvol'] = rvol
```

```python
data.head(5)
```

```python
rvol.head(5)
```

### Hidden Markov (Base)

```python
model = hmm.GaussianHMM(n_components=2, covariance_type='diag',
                        n_iter=1000)
```

```python
model.fit(np.column_stack([data['Returns'], rvol]))
```

```python
hidden_states = model.predict(np.column_stack([data['Returns']]))
```

```python
hidden_states
```

```python
df = pd.DataFrame(data['Returns'])
```

```python
df['State'] = hidden_states
```

```python
df['Price'] = data['Close']
```

```python
df.index = pd.to_datetime(df.index)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Price'], label='S&P 500')
plt.fill_between(df.index, df['Price'], color='r', where=(df['State']==1), alpha=0.3)
plt.fill_between(df.index, df['Price'], color='g', where=(df['State']==0), alpha=0.3)
plt.legend();
```

### Comparing the Model's Performance to the Underlying

```python
starting_balance = 10000
```

```python
df['Position'] = np.where(df['State'] == 1, 0, 1)
df['Cumret'] = ((1 + df['Returns']).cumprod() - 1)
df['Strategy'] = ((1 + df['Returns'] * df['Position']).cumprod() - 1)
df['ReturnsD'] = starting_balance  * (1 + df['Returns']).cumprod()
df['StrategyD'] = starting_balance * (1 + df['Strategy'])
```

```python
fig, ax1 = plt.subplots(2, 1, figsize=(10, 8))
plt.subplot(211)
plt.plot(df.index, df['Price'], label='S&P 500')
plt.fill_between(df.index, df['Price'], color='r', where=(df['State']==1), alpha=0.3)
plt.fill_between(df.index, df['Price'], color='g', where=(df['State']==0), alpha=0.3)
plt.legend()
plt.subplot(212)
plt.plot(df['Strategy'], label='HMM Strategy', color='m')
plt.plot(df['Cumret'], label='Long')
plt.ylabel('Returns')
plt.legend();
```

## Training/Testing Split

```python
split = int(len(data) * 0.7)
```

```python
train = data[:split]
test = data[split:]
```

```python
model = hmm.GaussianHMM(n_components=2, covariance_type='diag',
                        n_iter=10000)
```

```python
model.fit(np.column_stack((train['Returns'], train['rvol'])))
```

```python
testing = model.predict(np.column_stack([test['Returns']]))
```

```python
testing
```

```python
data_ = pd.DataFrame(test['Returns'])
```

```python
data_['Price'] = test['Close']
```

```python
data_['State'] = testing
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_.index, data_['Price'], label='S&P 500')
plt.fill_between(data_.index, data_['Price'], color='r', where=(data_['State']==1), alpha=0.3)
plt.fill_between(data_.index, data_['Price'], color='g', where=(data_['State']==0), alpha=0.3)
plt.legend();
```
