# Realized Volatility - Machine Learning Implementation

## Imports & Data Retrieval

We will be using some standard python packages (NumPy, Pandas, Matplotlib), hmmlearn (a python library that provides Hidden Markov Models), and a background script we put together as we were building this notebook, so the notebook doesn't get too cluttered. In the 'markov.py' script, we calculate our realized volatility, take care of data imports & processing, and some easy-to-use plotting functions. The fundamental question here is "Can we use machine learning while utilizing realized volatility calculations to generate signals regarding broad market conditions? 

```python
import markov as mk
import pandas as pd
from hmmlearn import hmm
import numpy as np
from pylab import mpl, plt
plt.style.use('seaborn-v0_8')
days = 10000 #10,000 Daily Bars
underlying = '^GSPC' #S&P 500
sp = mk.Markov(days, underlying) #Providing Ticker symbol and number of days to background script
data = sp.get_data() #Retrieving price data from script
rvol = sp.cc() #Retrieving realized volatility
data = data[29:] #Manually removing NaN's
data['rvol'] = rvol #Making the RVOL value a float
```

### Hidden Markov (Base)

* We'll start with running all of the in-sample data to determine the idea's viability.
* The model is fitted using the returns data & the realized volatility data. Our realized volatility calculation methodology is the Close-Close volatility estimator. 


```python
model = hmm.GaussianHMM(n_components=2, covariance_type='diag',
                        n_iter=1000) #Creating the model.
#Binary prediction (1 or 0, long or neutral/short), diagonal covariance, 1000 iterations
```

```python
model.fit(np.column_stack([data['Returns'], rvol])) 
# Fitting the model using realized vol & returns data. 
```

```python
prediction = model.predict(np.column_stack([data['Returns']]))
#Generating a returns prediction using the whole data set.
prediction
```

```python
df = pd.DataFrame(data['Returns'])
df['State'] = prediction
df['Price'] = data['Close']
df.index = pd.to_datetime(df.index)
# Creating new dataframe to analyze results
```

```python
mk.plot_signals(df)
```

**_Results_** 

* The model clearly looks pretty great.
* Without a training/testing split it's hard to determine if the model would have real-world viability.

### Comparing the Model's Performance to the Underlying

* The 'strategy' in question is either long or flat. 

```python
mk.plot_performance(df)
```

The **_real_** takeaway here is that *if* an allocator can avoid drawdowns, performance improves dramatically.

## Training/Testing Split

**_70% training, 30% Testing_**

```python
split = int(len(data) * 0.7) #Identifying the Split
train = data[:split]
test = data[split:]
model = hmm.GaussianHMM(n_components=2, covariance_type='diag',
                        n_iter=10000)
```

```python
model.fit(np.column_stack((train['Returns'], train['rvol'])))
```

```python
testing = model.predict(np.column_stack([test['Returns']]))
```

```python
data_ = pd.DataFrame(test['Returns'])
data_['Price'] = test['Close']
data_['State'] = testing
```

### Plotting the Testing Data

```python
mk.plot_signals(data_)
```

**_Looks pretty good!!_**

```python
mk.plot_performance(data_)
```
